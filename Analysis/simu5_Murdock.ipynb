```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../Data/simu5_Murdock.csv",header=None)
df
```

```python
df = df.pivot_table(index=2,columns=3,values=1).reset_index(drop=True)
df
```

```python
df['SD1'] = df[' SD+1'] - df[' Mean']
df['SD2'] = df[' Mean'] - df[' SD-1']
df['SD'] = df[['SD1','SD2']].mean(axis=1)
df['X'] = np.arange(0,6)
df['Y'] = df[' Mean']
df
```

```python
fig,ax = plt.subplots(figsize=(5,5))
ax.spines[['right', 'top']].set_visible(False)
ax.tick_params(axis="y",direction="in")
ax.tick_params(axis="x",direction="in")
plt.errorbar(x=df['X'],y=df['Y'],yerr=df['SD'], linewidth=2, marker='o', markersize = 5, capsize = 3, capthick = 2, linestyle='--')
plt.ylim([0,1])
plt.xlabel('Test Lag')
plt.ylabel('Probablility of Recall')
plt.savefig('fig/simu5_Murdock.pdf')
plt.show()
```
