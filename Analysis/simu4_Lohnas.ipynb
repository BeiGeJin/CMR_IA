```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
word_freq = [21,51,90,141,196,285,415,632,1163,4332]
```

```python
df_hr = pd.read_csv("../Data/simu4_Lohnas_HR.csv",header=None)
df_hr
```

```python
# x = df_far.loc[df_far[3]==' mean',0]
df_hr = df_hr.pivot_table(index=2,columns=3,values=1).reset_index(drop=True)
df_hr
```

```python
df_hr['SD1'] = df_hr[' sd+'] - df_hr[' mean']
df_hr['SD2'] = df_hr[' mean'] - df_hr[' sd-']
df_hr['SD'] = df_hr[['SD1','SD2']].mean(axis=1)
df_hr['X'] = word_freq
df_hr['Y'] = df_hr[' mean']
df_hr
```

```python
df_far = pd.read_csv("../Data/simu4_Lohnas_FAR.csv",header=None)
df_far
```

```python
# x = df_hr.loc[df_hr[3]==' mean',0]
df_far = df_far.pivot_table(index=2,columns=3,values=1).reset_index(drop=True)
df_far
```

```python
df_far['SD1'] = df_far[' sd+'] - df_far[' mean']
df_far['SD2'] = df_far[' mean'] - df_far[' sd-']
df_far['SD'] = df_far[['SD1','SD2']].mean(axis=1)
df_far['X'] = word_freq
df_far['Y'] = df_far[' mean']
df_far
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True,figsize=(5,6.5))
fig.subplots_adjust(hspace=0.03)

ax1.errorbar(x=df_hr['X'],y=df_hr['Y'],yerr=df_hr['SD'], linewidth=2, marker='o', markersize = 5, capsize = 3, capthick = 2, linestyle='--')
ax2.errorbar(x=df_far['X'],y=df_far['Y'],yerr=df_far['SD'], linewidth=2, marker='s', markersize = 5, capsize = 3, capthick = 2, linestyle='--')

# sns.lineplot(data=df_hr, y='Y', x='X', ax=ax1, marker = 'o', color = 'r', markersize = 8, linewidth = 2)
# sns.lineplot(data=df_far, y='Y', x='X', ax=ax2, marker = 'o', color = 'g', markersize = 8, linewidth = 2)

ax1.set_ylim(0.77, 0.95)
ax1.set_yticks(np.arange(0.80,0.96,0.05))
ax2.set_ylim(0.1, 0.28)
ax2.set_yticks(np.arange(0.1,0.30,0.05))
ax1.tick_params(axis="y",direction="in")
ax1.tick_params(axis="x",direction="in")
ax2.tick_params(axis="y",direction="in")
ax2.tick_params(axis="x",direction="in")
plt.xscale('log')
plt.xlim(5, 14000)

ax1.spines.bottom.set_visible(False)
ax1.spines.top.set_visible(False)
ax2.spines.top.set_visible(False)
ax1.spines.right.set_visible(False)
ax2.spines.right.set_visible(False)
ax1.tick_params(
    axis='x',
    which='both',
    bottom=False,
    top=False, 
    labelbottom=False,
    labeltop=False) 
ax2.minorticks_off()

d = .5  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax1.plot(0,0, transform=ax1.transAxes, **kwargs)
ax2.plot(0,1, transform=ax2.transAxes, **kwargs)

ax1.set_ylabel('hit rate')
ax1.set_xlabel(None)
ax2.set_ylabel('false alarm rate')
ax2.set_xlabel('word frequency')

plt.tight_layout(h_pad = 0)
plt.savefig('fig/simu4_Lohnas.pdf')
plt.show()
```
