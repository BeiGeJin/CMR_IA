```python
import pickle
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from optimization_utils import param_vec_to_dict
```

```python
file_list = glob.glob('outfiles_backup/S2_230805_100-100/*data*.pkl')
file_list = [i.replace("outfiles_backup/S2_230805_100-100/","") for i in file_list]
file_list = [i.replace(".pkl","") for i in file_list]
file_list = [tuple(map(int,i.split("data"))) for i in file_list]
file_list.sort()
file_list
```

```python
pso_df = pd.DataFrame(columns = ['iter','particle','err'])

for iter, p in file_list:
    path = "outfiles_backup/S2_230805_100-100/" + str(iter) + "data" + str(p) + ".pkl" 
    with open(path, 'rb') as inp:
        data = pickle.load(inp)
    try:
        err = data['err']
    except:
        err = np.nan
    pso_df = pd.concat([pso_df, pd.DataFrame({"iter":[iter],"particle":[p],"err":[err]})])
pso_df
```

```python
sns.scatterplot(data=pso_df, x="iter", y="err", hue="particle")
plt.ylim([0,0.04])
plt.show()
```
