```python
import pickle
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from optimization_utils import param_vec_to_dict
```

```python
from optimization_utils import make_boundary
_,_,what_to_fit = make_boundary(sim_name='S1')
what_to_fit
```

```python
file_list0 = glob.glob('outfiles_backup/S1_230830_500-200/*data*.pkl')
file_list0 = [i.replace("outfiles_backup/S1_230830_500-200/","") for i in file_list0]
file_list0 = [i.replace(".pkl","") for i in file_list0]
file_list0 = [tuple(map(int,i.split("data"))) for i in file_list0]
file_list0.sort()
```

```python
pso_df0 = pd.DataFrame(columns = ['iter','particle','err'] + what_to_fit)

for iter, p in file_list0:
    path = "outfiles_backup/S1_230830_500-200/" + str(iter) + "data" + str(p) + ".pkl" 
    with open(path, 'rb') as inp:
        data = pickle.load(inp)
    try:
        err = data['err']
        b = data['params']
        this_dist = {"iter":[iter],"particle":[p],"err":[err]}
        for i, w in enumerate(what_to_fit):
            this_dist[w] = [b[i]]
        pso_df0 = pd.concat([pso_df0, pd.DataFrame(this_dist)])
    except:
        pass
pso_df0
```

```python
file_list = glob.glob('outfiles_backup/S1_230908_500-200/*data*.pkl')
file_list = [i.replace("outfiles_backup/S1_230908_500-200/","") for i in file_list]
file_list = [i.replace(".pkl","") for i in file_list]
file_list = [tuple(map(int,i.split("data"))) for i in file_list]
file_list.sort()
```

```python
pso_df = pd.DataFrame(columns = ['iter','particle','err'] + what_to_fit)

for iter, p in file_list:
    path = "outfiles_backup/S1_230908_500-200/" + str(iter) + "data" + str(p) + ".pkl" 
    with open(path, 'rb') as inp:
        data = pickle.load(inp)
    try:
        err = data['err']
        b = data['params']
        this_dist = {"iter":[iter],"particle":[p],"err":[err]}
        for i, w in enumerate(what_to_fit):
            this_dist[w] = [b[i]]
        pso_df = pd.concat([pso_df, pd.DataFrame(this_dist)])
    except:
        pass
pso_df
```

```python
def get_min_row(df):
    return df.sort_values(by='err').head(1)

pso_df0['run'] = 0
pso_df['run'] = 1
tops0_df = pso_df0.groupby("particle").apply(get_min_row).reset_index(drop=True)
tops1_df = pso_df.groupby("particle").apply(get_min_row).reset_index(drop=True)
tops_df = pd.concat([tops0_df, tops1_df])
tops_df
```

```python
df_plot = pd.melt(tops_df, id_vars=['particle', 'run','err'], value_vars=what_to_fit, var_name='variables', value_name='value')
df_plot['color'] = df_plot.apply(lambda x: np.round(min((x['err']-0.005)//0.00005, 20)*0.00005+0.005, 4), axis=1)
df_plot
```

```python
df_plot1 = df_plot.loc[~np.isin(df_plot.variables,['phi_s','phi_d','omega'])]
```

```python
# plt.subplots(figsize=(20,8))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(28,16), sharex=True)
fig.subplots_adjust(hspace=0.05)  # adjust space between axes

# plot the same data on both axes
sns.stripplot(data=df_plot, x='variables', y='value', alpha=0.3, hue='color', palette='YlOrBr', ax=ax1)
sns.violinplot(data=df_plot, x='variables', y='value', width=3, color='lightgray', ax=ax1)
sns.stripplot(data=df_plot, x='variables', y='value', alpha=0.3, hue='color', palette='YlOrBr', ax=ax2)
sns.violinplot(data=df_plot, x='variables', y='value', width=3, color='lightgray', ax=ax2)

# zoom-in / limit the view to different portions of the data
ax1.set_ylim(2, 10)  # outliers only
ax2.set_ylim(-0.2, 1.2)  # most of the data

# hide the spines between ax and ax2
ax1.spines.bottom.set_visible(False)
ax2.spines.top.set_visible(False)
ax1.xaxis.tick_top()
ax1.tick_params(labeltop=False)  # don't put tick labels at the top
ax2.xaxis.tick_bottom()

# Now, let's turn towards the cut-out slanted lines.
# We create line objects in axes coordinates, in which (0,0), (0,1),
# (1,0), and (1,1) are the four corners of the axes.
# The slanted lines themselves are markers at those locations, such that the
# lines keep their angle and position, independent of the axes size or scale
# Finally, we need to disable clipping.

d = .5  # proportion of vertical to horizontal extent of the slanted line
kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
              linestyle="none", color='k', mec='k', mew=1, clip_on=False)
ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)

ax1.legend(title='MSE')
ax2.get_legend().set_visible(False)

ax1.set_title("Exp1 parameter distributions")
plt.savefig("S1_paramdist.png")
```
