```python
import pickle
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from optimization_utils import param_vec_to_dict
```

```python
from optimization_utils import make_boundary
_,_,what_to_fit = make_boundary(sim_name='S2')
what_to_fit
```

```python
file_list0 = glob.glob('outfiles_backup/S2_230821_500-200/*data*.pkl')
file_list0 = [i.replace("outfiles_backup/S2_230821_500-200/","") for i in file_list0]
file_list0 = [i.replace(".pkl","") for i in file_list0]
file_list0 = [tuple(map(int,i.split("data"))) for i in file_list0]
file_list0.sort()
```

```python
pso_df0 = pd.DataFrame(columns = ['iter','particle','err'] + what_to_fit)

for iter, p in file_list0:
    path = "outfiles_backup/S2_230821_500-200/" + str(iter) + "data" + str(p) + ".pkl" 
    with open(path, 'rb') as inp:
        data = pickle.load(inp)
    try:
        err = data['err']
        b = data['params']
        this_dist = {"iter":[iter],"particle":[p],"err":[err]}
        for i, w in enumerate(what_to_fit):
            this_dist[w] = [b[i]]
        pso_df0 = pd.concat([pso_df0, pd.DataFrame(this_dist)])
    except:
        pass
pso_df0
```

```python
file_list = glob.glob('outfiles_backup/S2_230912_500-200/*data*.pkl')
file_list = [i.replace("outfiles_backup/S2_230912_500-200/","") for i in file_list]
file_list = [i.replace(".pkl","") for i in file_list]
file_list = [tuple(map(int,i.split("data"))) for i in file_list]
file_list.sort()
```

```python
pso_df = pd.DataFrame(columns = ['iter','particle','err'] + what_to_fit)

for iter, p in file_list:
    path = "outfiles_backup/S2_230912_500-200/" + str(iter) + "data" + str(p) + ".pkl" 
    with open(path, 'rb') as inp:
        data = pickle.load(inp)
    try:
        err = data['err']
        b = data['params']
        this_dist = {"iter":[iter],"particle":[p],"err":[err]}
        for i, w in enumerate(what_to_fit):
            this_dist[w] = [b[i]]
        pso_df = pd.concat([pso_df, pd.DataFrame(this_dist)])
    except:
        pass
pso_df
```

```python
def get_min_row(df):
    return df.sort_values(by='err').head(1)

pso_df0['run'] = 0
pso_df['run'] = 1
tops0_df = pso_df0.groupby("particle").apply(get_min_row).reset_index(drop=True)
tops1_df = pso_df.groupby("particle").apply(get_min_row).reset_index(drop=True)
tops_df = pd.concat([tops0_df, tops1_df])
tops_df
```

```python
df_plot = pd.melt(tops_df, id_vars=['particle', 'run','err'], value_vars=what_to_fit, var_name='variables', value_name='value')
df_plot['color'] = df_plot.apply(lambda x: np.round(min((x['err']-0.002)//0.0001, 15)*0.0001+0.002, 4), axis=1)
df_plot
```

```python
plt.subplots(figsize=(16,8))
sns.stripplot(data=df_plot, x='variables', y='value', alpha=0.5, hue='color', palette='YlOrBr')
sns.violinplot(data=df_plot, x='variables', y='value', width=2, color='lightgray')
plt.legend(title='MSE')
plt.title("Exp2 parameter distributions")
plt.savefig("S2_paramdist.png")
```

```python
df = pd.concat([pso_df0, pso_df])
df.reset_index(inplace=True, drop=True)
df
```
